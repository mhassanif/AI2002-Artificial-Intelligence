```python
import random

# [weight, value]
items = [
    [1, 14],
    [3, 23],
    [7, 8],
    [4, 9],
    [5, 17],
    [6, 15],
]

knapsack_capacity = 10
pop_size = 8
max_gen = 5

# 1. Initialization
def generate_population():
    pop = []
    for i in range(pop_size):
        chromosome = []
        for i in range(len(items)):
            chromosome.append(random.randint(0, 1))
        pop.append(chromosome)
    return pop

# 2. Fitness Function
def calculate_fitness(chromosome):
    total_weight = 0
    total_value = 0
    for i in range(len(items)):
        total_weight += items[i][0] * chromosome[i]
        total_value += items[i][1] * chromosome[i]

    if total_weight > knapsack_capacity:
        # penalize capacity exceeding
        return total_value - total_weight
    else:
        return total_value

# 3. Selection
def selection(population, fitness_scores):
    population_with_fitness = []
    for i in range(len(population)):
        population_with_fitness.append((population[i], fitness_scores[i]))

    # Sort decending fitness
    population_with_fitness.sort(key=lambda item: item[1], reverse=True)
    print("Population with fitness : \n",population_with_fitness)

    # Select the top 2
    selected = [population_with_fitness[0][0], population_with_fitness[1][0]]
    return selected

# 4. Crossover
def single_point_crossover(parent1, parent2):
    crossover_point = random.randint(1, len(items) - 1)

    offspring1 = []
    offspring2 = []

    for i in range(crossover_point):
        offspring1.append(parent1[i])
        offspring2.append(parent2[i])

    for i in range(crossover_point, len(items)):
        offspring1.append(parent2[i])
        offspring2.append(parent1[i])

    return offspring1, offspring2

# 5. Mutation
def mutate(chromosome):
    for i in range(len(chromosome)):
        if random.random() < 0.1:
            # flip bit
            if chromosome[i] == 0:
                chromosome[i] = 1
            else:
                chromosome[i] = 0
    return chromosome


# 6. Replacement
def replace_population(population, offspring, fitness_scores):
    exceeding_capacity = []
    # Check those who exceed capacity
    for i in range(len(population)):
        score = calculate_fitness(population[i])
        if score < 0:
            exceeding_capacity.append(population[i])

    # Sort by ascending fitness
    sorted_population = [(chromosome, calculate_fitness(chromosome)) for chromosome in population]
    sorted_population.sort(key=lambda item: item[1])  # Ascending order

    if exceeding_capacity:
        for i in range(offspring.len()):
            sorted_population[i] = (offspring[i], calculate_fitness(offspring[i]))

        population = [x[0] for x in sorted_population]  # Extract
    else:
        # Replace least-fit with offspring
        for i in range(len(offspring)):
            sorted_population[i] = (offspring[i], calculate_fitness(offspring[i]))

        population = [x[0] for x in sorted_population]  # Extract

    return population

# Main Genetic Algorithm
population = generate_population()
for generation in range(max_gen):
    print("\n\nNEW ITERATION!\n")
    fitness_scores = []
    for chromosome in population:
        fitness_scores.append(calculate_fitness(chromosome))

    parents = selection(population, fitness_scores)
    print("Selected : \n",parents)

    offspring1, offspring2 = single_point_crossover(parents[0], parents[1])
    print("After Crossover : ")
    print(offspring1,offspring2)

    offspring1 = mutate(offspring1)
    offspring2 = mutate(offspring2)

    print("After Mutation : ")
    print(offspring1,offspring2)

    population = replace_population(population, [offspring1, offspring2], fitness_scores)

# Get the best solution
best_chromosome = None
best_fitness = 0

for chromosome in population:
    fitness = calculate_fitness(chromosome)
    if fitness > best_fitness:
        best_fitness = fitness
        best_chromosome = chromosome

print("Best chromosome:", best_chromosome)
print("Best fitness:", best_fitness)
```
